```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
with open("./text/AA/wiki_00", 'r') as f:
    data = f.read()

data
```

```python
sents = sent_tokenize(data)[1:][:-1]
sents
```

```python
sents[-1]
```

```python
total_sent = ""
def list2str(list_word):
    total_sent = ""
    for word in list_word:
        total_sent += (word + " ")
    return total_sent
    
for sent in sents:
    words = word_tokenize(sent)
    k = int(len(words)/2)
    
    pront_words = words[:k]
    back_words = words[k:]
    pront_words = list2str(pront_words)
    back_words = list2str(back_words)
    
    sent = pront_words + "\t" + back_words
    total_sent += (sent + "\n")
```

```python
total_sent
```

```python
with open("./BERT-pytorch/data/corpus.small", 'w') as f:
    f.write(total_sent)
```
